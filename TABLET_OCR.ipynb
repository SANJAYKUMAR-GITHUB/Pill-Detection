```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow

# Function to perform OCR on an image and extract tablet details
def ocr_tablet_details(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to enhance text visibility
    _, thresholded_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

    # Perform OCR using pytesseract
    text = pytesseract.image_to_string(thresholded_image)

    # Display the original image and the detected text
    print("Original Image:")
    cv2_imshow(image)
    print("\nDetected Text:")
    print(text)

# Replace 'your_image_path.png' with the path to your image file
image_path = 'Tab_2.png'
ocr_tablet_details(image_path)
```
